```python
# 4.4 Built-in Functions

# Python consists of lots of built-in functions. Eg. print() 
# Python also consists of different Modules, Libraries or Packages.
# These Modules, Libraries or Packages consists of lots of predefined functions for different topic areas, such as mathematics, plotting, handling database systems, etc.

# Python allows you to split your program into modules that can be reused in other Python programs. It comes with a large collection of standard modules that you can use as the basis of your programs.
# The Python standard Library consists of different modules for handling file I/O, basic mathematics, etc.
# The math module has all the basic math functions you need, such as: Trigonometric functions: sin(x), cos(x), etc.

# Example 4.4.1 Using the math module

# We create some basic examples how to use a Library, a Package or a Module.

#1 If we need only the sin() function, we can do like this:

from math import sin

x = 3.14

y = sin(x)
print(y)

print("--------------------------------------------------------------------")

#2 If we need a few functions, we can do like this:

from math import sin, cos

x = 3.14

y = sin(x)
print(y)

y = cos(x)
print(y)

print("----------------------------------------------------------------------")

#3 If we need many functions, we can do like this:

from math import *

x = 3.14

y = sin(x)
print(y)

y = cos(x)
print(y)

print("-----------------------------------------------------------------------")

#4 We can also use this alternative:

import math

x = 3.14
y = math.sin(x)

print(y)

print("-----------------------------------------------------------------------")

import math as mt

x = 3.14
y = mt.sin(x)

print(y)
```

```python
# Example 4.6.3: Creating Subplots

# The subplot command enables you to display multiple plots in the same window.

# Tying "subplot(m, n, p)" partitions the figure window into an m-by-n matrix
# of small subplots and selects the subpot for the current plot.

# The plots are numbered along the first row of the figure window, then the 
# second row, and so on.

import matplotlib.pyplot as plt
import numpy as np

xstart = 0
xstop = 2*np.pi
increment = 0.1

x = np.arange(xstart, xstop, increment)
y = np.sin(x)
z = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y, 'g')
plt.title('sin')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid()
plt.show()

plt.subplot(2, 1, 2)
plt.plot(x, z, 'r')
plt.title('cos')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid()
plt.show()
```

```python
# 4.6.4 Exercises

# Below you find different self-paced Exercises that you should go through and solve on your own.
# The only way to learn Python is to do lots of Exercises!

# Create sin(x) and cos(x) in 2 different plots.

# You should use all the Plotting functions listed below in your code:

# plot() # title() # xlabel() # ylabel() # axis() # grid() # legend() # show()

import matplotlib.pyplot as plt
import numpy as np

# Define the range of x-axis values
x = np.linspace(0.0, 5.0, 100)

# Calculate sine and cosine values
sine_y = np.sin(x)
cosine_y = np.cos(x)

# Create the first plot for sine function
plt.figure(figsize = (8, 5)) # Set the figure size
plt.plot(x, sine_y, label = 'sin(x)') # Plot the sine curve
plt.title('Sine Function') # Add a title
plt.xlabel('x') # Label the x-axis
plt.ylabel('sin(x)') # Label the y-axis
plt.axis('tight') # Set the axis limits to fit the data
plt.grid(True) # Add grid lines
plt.legend() # Add a legend

# Create the second plot for cosine function (in a separate figure)
plt.figure(figsize = (8, 5)) # Set the figure size
plt.plot(x, cosine_y, label = 'cos(x)') # Plot the sine curve
plt.title('Cosine Function') # Add a title
plt.xlabel('x') # Label the x-axis
plt.ylabel('cos(x)') # Label the y-axis
plt.axis('tight') # Set the axis limits to fit the data
plt.grid(True) # Add grid lines
plt.legend() # Add a legend

# Display both plots
plt.show()
```

```python
# 5.2 Arrays

# An array is a special variable, which can hold more than one value at a time.

# 5.2.1 Using For loops in Python

data = [1.6, 3.4, 5.5, 9.4]

N = len(data)

print("---------------------------------------------------------------------")

print(N)

print("---------------------------------------------------------------------")

print(data[2])

print("---------------------------------------------------------------------")

for x in data:
    print(x)

print("---------------------------------------------------------------------")

data.append(11.4)

N = len(data)

print(N)

print("---------------------------------------------------------------------")

for x in data:
    print(x)
```

```python
# You can use Arrays in Loops

data = [1.6, 3.4, 5.5, 9.4]

for x in data:
    print(x)

print("-----------------------------------------------------------------------")


# You can return the number of elements in the array

N = len(data)

print(N)

print("-----------------------------------------------------------------------")


data.append(11.4)

print(data)

print("-----------------------------------------------------------------------")


# Original data list

data = [1.6, 3.4, 5.5, 9.4, 11.4]

# sort()

data.sort() # this sorts the list in place in ascending order

print("Sorted list:", data)

print("-----------------------------------------------------------------------")


# To sort in descending order

data.sort(reverse = True) # sorts the list in descending order

print("Sorted list in descending order:", data) 

print("-----------------------------------------------------------------------")


# clear()

data.clear() # this empties the list, removing all its elements

print("Cleared list:", data)

print("-----------------------------------------------------------------------")


# Reset data to original for next operations

data = [1.6, 3.4, 5.5, 9.4, 11.4]

print(data)

print("-----------------------------------------------------------------------")


N = len(data)

print(N)

print("-----------------------------------------------------------------------")


# copy()

data_copy = data.copy() # changes to 'data_copy' won't affect the original 'data'

print("Copied list: ", data_copy)

print("-----------------------------------------------------------------------")


# count(value)

value_to_count = 5.5 

count_value = data.count(value_to_count) # This counts how many times the value '5.5' appears in the list

print(f"Count of {value_to_count} in the list: ", count_value)

print("-----------------------------------------------------------------------")


# insert(index, value)

index_to_insert = 2

value_to_insert = 7.7

data.insert(index_to_insert, value_to_insert) # inserts the value '7.7' at the index '2'

print("List after insertion: ", data)

print("-----------------------------------------------------------------------")


# remove(value)

value_to_remove = 7.7

data.remove(value_to_remove) # removes the first occurence of the value '7.7' from the list. If the value doesn't exist, it raises a 'ValueError'

print("List after removal: ", data)

print("-----------------------------------------------------------------------")


cars = ["Volvo", "Tesla", "Ford"]

index = 2

x = cars[index]

print(x)

print("-----------------------------------------------------------------------")


```

```python
# Example 5.3.2 Using For Loops for Summation of Data

# You typically want to use a For loop to find the sum of a given data set.

data = [1, 5, 6, 3, 12, 3]

sum = 0

# Find the sum of all the numbers

for x in data:
    sum = sum + x

print(sum)

# Find the Mean or Average of all the numbers

N = len(data)

mean = sum/ N

print(mean)
```
